```python
import matplotlib.pyplot as plt
```

```python
L = 50
M1 = L/4
M2 = 3*M1
dt = 0.25
dx = 1.0
D = 1.0
k = D*dt/(dx**2)

tl = []
tmax = 1
```

```python
i = 0
f = []


while i < L:
    if i > M1 and i < M2:
        f.append(1.0) 
    else:
        f.append(0.0)
    i+=1
print(len(f))
t = 0
fl = [f]
while t < 1000:
    i = 1
    g = []
    g.append(0)
    tl.append(0)

    while i < L-1:
        g.append(f[i] + k*(f[i-1] - 2*f[i] + f[i+1]))
        i+=1
    g.append(0)

    f = g
    tl.append(t)
    t+=dt
    if t%1 == 0:
        fl.append(f)
len(fl)
```

```python
plt.plot(range(len(fl[0])), fl[0])
plt.plot(range(len(fl[10])), fl[10])
plt.plot(range(len(fl[50])), fl[50])
plt.plot(range(len(fl[100])), fl[100])
plt.plot(range(len(fl[500])), fl[500])
plt.plot(range(len(fl[1000])), fl[1000])
```
