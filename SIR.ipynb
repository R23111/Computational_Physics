O método SIR parte de 3 princípios relativamente simples: 


1.   O número de sucetíveis a doença (S) diminui  numa relação direta com o numero de pessoas que se infetam (I)
2.   O número de infectados (I) será uma taxa, chamada aqui de $\beta$, do número de sucetíveis, e atenuada com o número de de pessoas já "pré" infectadas
3.   O número de recuperados (R) será uma taxa, chamada de $\gamma$, do total de infectados.

Ou seja, a cada passagem de tempo, o número de infectados aumentará $\beta * S * I$, mas diminuirá os que se recuperarem  ($\gamma$ * I). Com este pensamento é fácil deduzir um sitema de equações a serem resolvidas:

$$\frac{dS}{dt}=-\beta I S$$


$$\frac{dI}{dt}=\beta I S - \gamma I$$


$$\frac{dR}{dt}=\gamma I $$

um método numérico é facilmente obtido a partir destas equações:

$$S = S-\beta IS\Delta t$$
$$I = I + \beta IS\Delta t - \gamma I \Delta t$$
$$ R = R + \gamma I \Delta t$$




```
import numpy as np
from matplotlib import pyplot as plt

y = 0.5
binit = [0.1, 0.012, 0.006]
dt = 0.01
tfl = [10, 20, 50]
j = 0



for b in binit:
    s = 99
    i = 1
    r = 0
    t = 0
    tf = tfl[j]
    s_list, i_list, r_list, t_list = [], [], [], []

    while t < tf:
        t = t + dt
        s = s - b * i * s * dt
        i = i + (b * i * s *dt) - (y * i * dt)
        r = r + y * i * dt

        s_list.append(s)
        t_list.append(t)
        i_list.append(i)
        r_list.append(r)

    plt.plot(t_list, s_list)
    plt.plot(t_list, i_list)
    plt.plot(t_list, r_list)
    plt.legend(["suscetíveis", "infectados", "recuperados"])
    j += 1
    plt.show()


```

Comparando os dois primeiros gráficos, observa-se que, apesar de o número final de infectados ser o mesmo, por ele ocorrer curante um período de tempo maior, faz que o maior número simultâneo de pessoas com a doença seja menor.
O que os gráficos e as equações mostram, também, é a importância de diminuir o número de pessoas infectadas simultâneamente, pois o coeficiente de infecção $\beta$ é amplificado por este número.
