#Método de Euler 
Queremos resolver 
$$\frac{dx}{dt}=f(x)\;,$$ com $x(0)=x_0$. 
##Explícito
$$x_{i+1}=x_i+f(x_i)\Delta t$$
iniciando com $x_1=x_0$ obtém-se por sucessivas iterações $x_1,x_2,\ldots,x_n $.
##Implícito
$$x_{i+1}=x_i+f(x_{i+1})\Delta t$$
se for possível isolar $x_{i+1}$ à esquerda da igualdade, pode-se proceder por iterações como no caso anterior.
#Exemplo: Decaimento Radioativo
Trataremos um exemplo com solução analítica para poder verificar a soluções numéricas. Além disso, por ser uma equação linear pode-se integrar também pelo método implícito.

$$\frac{dN}{dt}=-N/\tau\;,$$ com $N(0)=N_0$.

Solução Analítica:
$$N=N_0e^{-\frac{t}{\tau}} $$ 

Euler explícito: 
$$ N_{i+1}=N_i-\frac{N_i}{\tau} \Delta t$$
$$ N_{i+1}=N_i(1-\frac{\Delta t}{\tau})$$

Euler implícito: 
$$ N_{i+1}=N_i-\frac{N_{i+1}}{\tau} \Delta t$$
$$ N_{i+1}=N_i(1+\frac{\Delta t}{\tau})$$

```
#programa em python3 
import numpy as np
from matplotlib import pyplot as plt
N0,dt,tau,tf,t=10,0.5,2,5,0 #entrada
t_list,N_list,Nan_list, Nimp_list=[0],[N0],[N0],[N0]
N=N0
Nimp=N0
while t<tf :
  N = N*(1-dt/tau)
  t+=dt
  Nan=N0*np.exp(-t/tau)
  Nimp=Nimp/(1+(dt/tau))
  Nimp_list.append(Nimp)
  N_list.append(N)
  t_list.append(t)
  Nan_list.append(Nan)
plt.scatter(t_list,N_list)
plt.title('Metodo de Euler Explicito')
plt.xlabel('time')
plt.ylabel('Number of Radioactive Atoms')
plt.plot(t_list,Nan_list)
plt.plot(t_list,Nimp_list)
#print(t_list)
#print(N_list)
plt.show()
```
