# Predador-Presa

```
import matplotlib.pyplot as plt
```

```
# Euler-Cromer "Euler Errado"

def loop(n,p, a, tf,dt):
  t = 0
  nl, pl, tl = [],[],[]
  while(t <= tf):
    n += n*(1-p) *dt
    ## nn = n + n*(1-p) *dt   --> Euler "Certo"
    p += a * p * (n-1) * dt
    ## n = nn   --> Euler Certo
    nl.append(n)
    tl.append(t)
    pl.append(p)
    t += dt
  return [tl, nl, pl]
```

GÅ•afico $P $x$ N$

```
nl = [1, 0.5, 0.2, 0.1, 0.05]
for n in nl:
  g = loop(n=n, p=0.8, a=0.5, tf=100, dt=0.01)
  plt.plot(g[1],g[2])
  plt.legend(nl)
  plt.xlabel("n")
  plt.ylabel("p")
  plt.title("Predador presa para P_0 = 0.8")
```

GrÃ¡ficos $P $x$ t$ e $N $x$ t$

```
g = loop(n=0.1, p=0.8, a=0.5, tf=45, dt=0.01)
plt.plot(g[0],g[1])
plt.legend("n")
plt.plot(g[0],g[2])
plt.legend(["n","p"])
plt.xlabel("t")
plt.ylabel("n, p")
plt.title("Predador presa para P_0 = 0.8, N_0 = 0.1, alpha = 0.5")
```
