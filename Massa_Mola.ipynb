# Método de Euler - Massa-Mola

---

### Problema geral da mecânica clásscia (Newton)

$$ m \frac{d^2x}{dt^2} = f(x,v,t) $$


Equação diferencial de 2ª ordem; o método de euler serve para equações de 1ª ordem.

Equações de ordem n podem ser transformadas em n eq. de 1ª ordem:

$$ m \frac{dv}{dt} = f(x,v,t) $$

$$ \frac{dx}{dt} = v(t) $$


---


### Método de Euler - Eq. Newton 1-D

$$ \frac{dv}{dt} = a(t) $$

$$ \frac{dx}{dt} = v(t) $$


#### Aplicando o método de Euler:

$$ v (t+\Delta t) = v(t) + a(t)\Delta t$$

$$ x(t + \Delta t) = x(t) + v(t) \Delta t $$

Particula massa $m$m, mola constante $k$

$$m \frac{d^2x}{dt^2} = - \frac{k}{m}x$$

$$ \frac{dv}{dt} = a(t) = -\omega^2 x $$

$$ \frac{dx}{dt} = v(t) $$

sendo 

$\omega = \sqrt{\frac{k}{m}}$


---

### Massa-Mola (Euler)

$$ x(x + \Delta t) = x(t) + v(t)\Delta t $$

$$ v(t+\Delta t) = v(t) - \omega^2 x(t)\Delta t$$

```
import matplotlib.pyplot as plt
from math import *
```

```
def a(x, w2):
    ac = -w2*x
    return ac

def loop(x, v, w2, dt):
  tau = 2*pi/sqrt(w2)
  tf = 5*tau
  t = 0
  vl, xl, tl = [],[],[]

  while(t <= tf):
    vt = v + a(x, w2) * dt
    x += v * dt
    v = vt
    
    tl.append(t)
    vl.append(v)
    xl.append(x)
    
    t += dt 

  
  return [tl,xl,vl]
```

```
g = loop(x=1., v=0., w2=1., dt = (2*pi/180))


for i in range(1, len(g)):
  plt.plot(g[0],g[i])
```
