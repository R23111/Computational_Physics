# MÃ©todo de Verlet

```python
import numpy as np
from matplotlib import pyplot as plt

w = 1
g = 10
lng = 10
tf = 30
m = 1
t0 = 0


xinit = [1, 1.5, 3, 9.42477, -9.42477]

fig, axs = plt.subplots(2, 2)
fig.suptitle('MÃ©todo de Verlet')


for x0 in xinit:
    t = t0
    v0 = 0
    if x0 == 9.42477:
        v0 = -0.2
    if x0 == -9.42477:
        v0 = 0.2
    dt = 0.1

    x_list = []
    t_list = []
    v_list = []
    et_list = []

    x1 = x0 + v0 * dt

    while (t < tf):
        t = t + dt
        x2 = 2 * x1 - x0 - (w ** 2) * np.sin(x1) * dt ** 2
        v1 = (x2 - x0) / (2 * dt)

        Ec = m * (v1 * v1) * lng**2 * 0.5
        Ep = m * g * lng * (1 - np.cos(x1))
        Et = Ec + Ep

        x_list.append(x1)
        v_list.append(v1)
        t_list.append(t)
        et_list.append(Et)

        x0 = x1
        x1 = x2

    axs[0, 0].plot(t_list, x_list)
    axs[0, 0].set_title("Î¸ x t")

    axs[1, 0].plot(x_list, v_list)
    axs[1, 0].set_title("espaÃ§o de fases (v x Î¸)")

    axs[0, 1].plot(t_list, v_list)
    axs[0, 1].set_title("v x t")

    axs[1, 1].plot(t_list, et_list)
    axs[1, 1].set_title("Et x t")


for i in range(2):
    for j in range(2):
        axs[i, j].legend(["Î¸0 = 1.0", "Î¸0 = 1.5", "Î¸0 = 3",
                    "Î¸0 = 9.42477", "Î¸0 = -9.42477"])


plt.show()
```
